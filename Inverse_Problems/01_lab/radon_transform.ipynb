## Реализация

### Библиотеки

```python
import numpy as np
from PIL import Image
from skimage.transform import radon, iradon
```

### Исходное изображение в grayscale

```python
img_gray = Image.open("initial.jpg").convert("L")
img_gray
```

### Преобразование в формат с плавающей запятой

Важно, т.к. иначе целочисленные значения просто уйдут в 0

```python
img_array = np.asarray(img_gray).astype(np.float64)
```

### Задаём количество углов поворота

Шаг по углам

```python
step = 10
#step = 180/max(img_array.shape)  # чтобы получить исходную картинку
```

Брать всю картинку или только круг в центре

```python
circle = False  
```

Углы поворота

```python
theta = np.arange(0.0, 180.0, step)
```

### Прямое преобразование Радона

```python
img_radon = radon(img_array, theta=theta, circle=circle)
```

```python
Image.fromarray(img_radon).convert("L")
```

### Обратное преобразование Радона

```python
img_reconstructed = iradon(img_radon, theta=theta, circle=circle)
```

```python
Image.fromarray(img_reconstructed).convert("L")
```
